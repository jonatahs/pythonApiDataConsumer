```python
# Imports
import json
from urllib import response
import requests
import pandas as pd
from pandas.io.json import json_normalize
import flatten_json
from apiKey import apiKey
```

```python
# url with apikey 
url = apiKey
# api connection
response = requests.get(url)
# print connection api, if print 200, the api its working!
print(f'Reponse API:{response}')

print('-----------------------------')
print('\n')
print('Datas from API')
# reading datas from api
data = response.json()
#print json
print(data)
```

```python
# Select articles in json
articlesList = data['articles']
articles = [i for i in articlesList]
```

```python
doiValue = [doi.get('doi') for doi in articles]
titleValue = [title.get('title') for title in articles]
publisherValue = [publisher.get('publisher') for publisher in articles]
pdfValue = [pdf.get('pdf_url') for pdf in articles]
abstractValue = [abstract.get('abstract') for abstract in articles]
rankValue = [rank.get('rank') for rank in articles]
```

```python
articlesdf = pd.DataFrame({'doi':doiValue,'title':titleValue,'rank':rankValue,'pdf_url':pdfValue,'abstract':abstractValue,'publisher':publisherValue})
articlesdf
```

```python
articlesdf.to_parquet('IEEE.parquet',index=False)
```

```python
articlesdf.to_csv('IEEE.csv',index=False)
```
